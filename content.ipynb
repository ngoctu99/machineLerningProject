```python
import numpy
from scipy import stats
# given a list speed of auto, then using mean() method to find the average speed

speedAuto = [99,86,87,88,111,86,103,87,94,78,77,85,86]

averageSpeed = numpy.mean(speedAuto)
print(f"Average speed: {averageSpeed}")


# it is important sorting the list before you find median
# if the number of item list ungerade then median / 2
medianSpeed = numpy.median(speedAuto)
print(f"Median Speed: {medianSpeed}")


# mode() using to pick the item which most appears

modeSpeed = stats.mode(medianSpeed);
```

```python
# Standard deviation (do lech chuan) is a number that describes how spread out (phan tan) the values are
#A low standard deviation means that most of the numbers are close to the mean (average) value.
#A high standard deviation means that the values are spread out over a wider range.
import numpy

speedAuto = [86,87,88,86,87,85,86]

averageSpeed = numpy.mean(speedAuto)
print(f"medium: {speedAuto}")


# method std() of numpy to get standard deviation

standardDeviation = numpy.std(speedAuto)
print(f"standard deviation: {standardDeviation}")
```

```python
import numpy
import math

speedAuto = [32,111,138,28,59,77,97]

varianceValue = numpy.var(speedAuto)
print(f"variance: {varianceValue}")

squareRoot = math.sqrt(varianceValue);
print(squareRoot)

averageSpeed = numpy.std(speedAuto)
print(f"average speed: {averageSpeed}")
```

```python
#Percentiles are used in statistics to give you a number that describes the value that a given percent of the values are lower than.

import numpy

ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

percentile = numpy.percentile(ages, 75);
print(f"percentile: {percentile}")

# the result is 43.0, that mean 75 percent people are 43 years old or younger


percentile90 = numpy.percentile(ages, 90)
print(f"percentile 90: {percentile90}")

# the result is 61.0, that mean 90 percent people are 43 years old or younger
```

```python
import numpy
import matplotlib.pyplot as plt

# Normal Data distribution 
# numpy.random.normal(mean, standard devitation, 100000 values)
dataSet = numpy.random.normal(5.0 , 1.0, 100000)

plt.hist(dataSet, 100, edgecolor = "black");
plt.show()
```
